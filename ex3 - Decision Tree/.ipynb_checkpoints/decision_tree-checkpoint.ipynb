```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import *
from sklearn.metrics import precision_recall_fscore_support
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
df = pd.read_csv('Admission_Predict.csv')
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
threshold = 0.90
df.loc[df['Chance of Admit '] > threshold, 'Chance of Admit '] = 1
df.loc[df['Chance of Admit '] <= threshold, 'Chance of Admit '] = 0
```

```python
X = df.drop(['Serial No.', 'Chance of Admit '], axis = 1)
y = df['Chance of Admit ']
```

```python
(train_X, test_X, train_y, test_y) = train_test_split(X, y, random_state=2, test_size=0.2)
```

```python
model = DecisionTreeClassifier()
model.fit(train_X, train_y)
pred_y = model.predict(test_X)
```

```python
print('Classification Report =\n',classification_report(test_y, pred_y))
print('Accuracy =', accuracy_score(test_y, pred_y))
```

```python
ConfusionMatrixDisplay(confusion_matrix(test_y, pred_y), display_labels= ['Admitted', 'Not Admitted']).plot()
```

```python
f_name = ['GRE Score','TOEFL Score','University Rating','SOP','LOR','CGPA','Research']
plt.figure(figsize=(30,20))
img = plot_tree(model, filled=True, feature_names=f_name, class_names=[str(x) for x in model.classes_])
```
